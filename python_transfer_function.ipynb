(from State Space and Transfer Functions in Python : https://www.youtube.com/watch?v=2lQbGQ_cQ3w)  

$$\tau \frac{dy}{dt} = -y + K_p u$$

**Let's try first by defining a transfer function**

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import signal
from scipy.integrate import odeint
import numpy as np

Kp = 3
tau = 2
num = [Kp]
den = [tau, 1]
sys = signal.TransferFunction(num,den)
```

```python
t,y = signal.step(sys)
plt.figure(1)
plt.plot(t,y,'r');
```

**Then by using an ODE integrator**

```python
def model(y,t):
    u = 1
    return(-y + Kp*u)/tau

t = np.linspace(0,14,100)
y = odeint(model,0,t)
plt.plot(t,y);
```

**Then by defining a state-space equation**  
The same system can be rewritten as :  
$$\dot{x} = \frac{-1}{\tau}x + \frac{K_p}{\tau}u$$
and
$$y = x$$



```python
A = -1.0
B = Kp/tau
C = 1.0
D = 0.0

sys = signal.StateSpace(A,B,C,D)
t,y = signal.step(sys)
plt.plot(t,y);
```
